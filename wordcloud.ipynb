```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from collections import Counter
from konlpy.tag import Okt
from PIL import Image
import numpy as np
```

### 키워드

* 부산 전체

```python
path = "C:\\Users\\82108\\key\\"
file_list = os.listdir(path)
file_list_py = [file for file in file_list if file.endswith('.csv')] 

key_df = pd.DataFrame()
for i in file_list_py:
    data = pd.read_csv(path + i, encoding = "CP949")
    key_df = pd.concat([key_df,data])
    
key_df = key_df.reset_index(drop = True)
key_df
```

```python
key_df.info()
```

```python
kw = key_df['여행유형/트렌드 관련 키워드'].str.split("_")

key1 = pd.DataFrame(kw.str.get(0))
key2 = pd.DataFrame(kw.str.get(1))
key3 = pd.DataFrame(kw.str.get(2))

result = pd.concat([key1, key2, key3], ignore_index = True).dropna(axis = 0)
```

```python
result.reset_index(inplace = True, drop = True)
result
```

```python
df = result['여행유형/트렌드 관련 키워드'].to_list()
df
```

```python
wordCount = {} 

for word in df:
    wordCount[word] = wordCount.get(word, 0) + 1 
    keys = sorted(wordCount.keys())

for word in keys:
    print(word + ':' + str(wordCount[word]))
```

```python
wc = WordCloud(font_path='C:\\Users\\82108\\AppData\\Local\\Microsoft\\Windows\\Fonts\\DXMSubtitlesM-KSCpc-EUC-H.ttf', #폰트
               background_color='white',                              #배경색
               width=800, height=600 ,                                #사이즈설정
               max_words=200)                                         #단어갯수
cloud = wc.generate_from_frequencies(wordCount)                            #사전형태 데이터

plt.figure(figsize=(10,10))                                           #액자사이즈설정
plt.axis('off')                                                       #테두리 선 없애기
plt.imshow(cloud,interpolation="bilinear")
```

* 관광활성화 지수 top5

```python
path = "C:\\Users\\82108\\key5\\"
file_list = os.listdir(path)
file_list_py = [file for file in file_list if file.endswith('.csv')] 

key_df = pd.DataFrame()
for i in file_list_py:
    data = pd.read_csv(path + i, encoding = "CP949")
    key_df = pd.concat([key_df,data])
    
key_df = key_df.reset_index(drop = True)

kw = key_df['여행유형/트렌드 관련 키워드'].str.split("_")

key1 = pd.DataFrame(kw.str.get(0))
key2 = pd.DataFrame(kw.str.get(1))
key3 = pd.DataFrame(kw.str.get(2))

result = pd.concat([key1, key2, key3], ignore_index = True).dropna(axis = 0)
result.reset_index(inplace = True, drop = True)
result = result['여행유형/트렌드 관련 키워드'].to_list()

wordCount = {} 

for word in result:
    wordCount[word] = wordCount.get(word, 0) + 1 
    keys = sorted(wordCount.keys())

wc = WordCloud(font_path='C:\\Users\\82108\\AppData\\Local\\Microsoft\\Windows\\Fonts\\DXMSubtitlesM-KSCpc-EUC-H.ttf', #폰트
               background_color='white',                              #배경색
               width=800, height=600 ,                                #사이즈설정
               max_words=200)                                         #단어갯수
cloud = wc.generate_from_frequencies(wordCount)                            #사전형태 데이터

plt.figure(figsize=(10,10))                                           #액자사이즈설정
plt.axis('off')                                                       #테두리 선 없애기
plt.imshow(cloud,interpolation="bilinear")
```
