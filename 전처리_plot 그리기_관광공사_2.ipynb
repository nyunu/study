```python
f = pd.read_csv("C:\\Users\\82108\\DataLab_data\\경상남도\\거제시\\SNS언급량\\2019.csv", encoding="cp949")
f.columns
```

```python
#
city = {"경기도":gyeonggi1, "경상남도":gyeongnam1, "경상북도":gyeongbook1, "전라남도":junnam1, "전라북도":junbook1,
       "충청남도":chungnam1, "충청북도":chungbook1, "강원도":gangwon1}
city
```

```python
folder_name = ["SNS언급량", "방문자수"]
```

```python
for do, si in city.items():
    for si_name in si:
        for fn in folder_name:
            f2019 = pd.read_csv("C:\\Users\\82108\\DataLab_data\\"+do+"\\"+si_name+"\\"+fn+"\\2019.csv",encoding='cp949')
            f2020 = pd.read_csv("C:\\Users\\82108\\DataLab_data\\"+do+"\\"+si_name+"\\"+fn+"\\2020.csv",encoding='cp949')
            f2021 = pd.read_csv("C:\\Users\\82108\\DataLab_data\\"+do+"\\"+si_name+"\\"+fn+"\\2021.csv",encoding='cp949')
            f2022 = pd.read_csv("C:\\Users\\82108\\DataLab_data\\"+do+"\\"+si_name+"\\"+fn+"\\2022.csv",encoding='cp949')
            
            f2019.drop(['지역명'], axis=1, inplace=True)
            f2020.drop(['지역명'], axis=1, inplace=True)
            f2021.drop(['지역명'], axis=1, inplace=True)
            f2022.drop(['지역명'], axis=1, inplace=True)
            
            result = pd.concat([f2019, f2020, f2021, f2022], ignore_index=True)

            result.to_csv('C:\\Users\\82108\\result_data\\' + do + '\\' + fn + '\\' + si_name +".csv", encoding='cp949', index=False)
```
